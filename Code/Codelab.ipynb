```python
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt

# Buka gambar yang diunggah
img = Image.open("../Asset/umm.png")  # ".." digunakan untuk naik satu folder

# Terapkan filter pertama (contoh: BLUR)
image_after = img.filter(ImageFilter.BLUR)

# Terapkan filter kedua (contoh: CONTOUR)
image_after2 = img.filter(ImageFilter.CONTOUR)

# Tampilkan hasil sebelum dan sesudah
plt.figure(figsize=(10, 5))

# Gambar asli
plt.subplot(1, 3, 1)
plt.title("Original")
plt.imshow(img)
plt.axis("off")

# Gambar dengan filter pertama
plt.subplot(1, 3, 2)
plt.title("Filter: BLUR")
plt.imshow(image_after)
plt.axis("off")

# Gambar dengan filter kedua
plt.subplot(1, 3, 3)
plt.title("Filter: CONTOUR")
plt.imshow(image_after2)
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image, ImageFilter

# --- Pastikan Folder dan File ---
# Path relatif untuk gambar dan output
img_path = "../Asset/umm.png"
output_folder = "../Output"
img = mpimg.imread(img_path)

# 1a. Membuat judul
plt.figure(figsize=(10, 5))
plt.subplot(2, 3, 1)
plt.title("Original Image")
plt.imshow(img)
plt.axis("on")  # Sumbu ditampilkan

# 1b. Menyembunyikan sumbu Axis
plt.subplot(2, 3, 2)
plt.title("Axis Hidden")
plt.imshow(img)
plt.axis("off")  # Sumbu disembunyikan

# 1c. Menambahkan teks dan anotasi
plt.subplot(2, 3, 3)
plt.title("Text & Annotation")
plt.imshow(img)
plt.axis("off")
plt.text(50, 50, "Hello, Matplotlib!", color="white", fontsize=12, backgroundcolor="black")

# 1d. Rotate gambar
rotated_img = Image.open(img_path).rotate(45, expand=True)  # Rotasi 45 derajat
plt.subplot(2, 3, 4)
plt.title("Rotated Image")
plt.imshow(rotated_img)
plt.axis("off")

# Menyimpan hasil rotasi
rotated_img_save_path = os.path.join(output_folder, "rotated_image.png")
rotated_img.save(rotated_img_save_path)
print(f"Gambar hasil rotasi disimpan di: {rotated_img_save_path}")

# 1f. Melakukan filtering
filtered_img = Image.open(img_path).filter(ImageFilter.CONTOUR)  # Terapkan filter
plt.subplot(2, 3, 5)
plt.title("Filtered Image")
plt.imshow(filtered_img)
plt.axis("off")

# Menampilkan hasil keseluruhan
plt.tight_layout()
plt.show()
```

```python
from PIL import Image
import os

# TODO 1: Path gambar
background_path = "../Asset/fungsio.png"  # Path gambar background
overlay_path = "../Asset/umm.png"        # Path gambar overlay
output_path = "../Output/result.png"     # Path untuk menyimpan hasil

# TODO 2: Buka gambar
background_image = Image.open(background_path)
overlay_image = Image.open(overlay_path)

# TODO 3: Konversi overlay ke mode RGBA untuk mendukung transparansi
overlay_image = overlay_image.convert("RGBA")
background_image = background_image.convert("RGBA")

# TODO 4: Resize gambar overlay (opsional)
overlay_image = overlay_image.resize((100, 100))  # Atur ukuran overlay 100x100 pixel

# Tentukan posisi koordinat untuk menempatkan overlay di tengah background
x_center = (background_image.width - overlay_image.width) // 2
y_center = (background_image.height - overlay_image.height) // 2

# TODO 5: Sisipkan gambar overlay ke dalam gambar background
background_image.paste(overlay_image, (x_center, y_center), overlay_image)  # Mask untuk transparansi

# TODO 6: Simpan gambar hasil edit
background_image.save(output_path, format="PNG")
print(f"Gambar hasil edit disimpan di: {output_path}")

# TODO 7: Tampilkan gambar hasil edit
background_image.show()
```
